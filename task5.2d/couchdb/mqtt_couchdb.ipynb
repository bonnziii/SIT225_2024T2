```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file generated from CouchDB data
df = pd.read_csv('gyroscope_data_couchdb.csv')

# Display the first few rows of the data to verify the structure
print(df.head())

# Plot 'x' variable
plt.figure()
plt.plot(df['x'], label='X Axis', color='blue')
plt.title('Gyroscope X Axis')
plt.xlabel('Time (samples)')
plt.ylabel('X Value')
plt.legend()
plt.savefig('gyroscope_x_axis_couchdb.png')  # Save the plot as PNG
plt.show()

# Plot 'y' variable
plt.figure()
plt.plot(df['y'], label='Y Axis', color='orange')
plt.title('Gyroscope Y Axis')
plt.xlabel('Time (samples)')
plt.ylabel('Y Value')
plt.legend()
plt.savefig('gyroscope_y_axis_couchdb.png')  # Save the plot as PNG
plt.show()

# Plot 'z' variable
plt.figure()
plt.plot(df['z'], label='Z Axis', color='green')
plt.title('Gyroscope Z Axis')
plt.xlabel('Time (samples)')
plt.ylabel('Z Value')
plt.legend()
plt.savefig('gyroscope_z_axis_couchdb.png')  # Save the plot as PNG
plt.show()

# Plot all three variables on a single graph
plt.figure()
plt.plot(df['x'], label='X Axis', color='blue')
plt.plot(df['y'], label='Y Axis', color='orange')
plt.plot(df['z'], label='Z Axis', color='green')
plt.title('Gyroscope X, Y, Z Axes Combined')
plt.xlabel('Time (samples)')
plt.ylabel('Value')
plt.legend()
plt.savefig('gyroscope_xyz_combined_couchdb.png')  # Save the combined plot as PNG
plt.show()
```
