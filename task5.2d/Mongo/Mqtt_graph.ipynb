```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gyroscope_data.csv')

# Display the first few rows to verify the structure
df.head()
```

```python
# Plot 'x' variable
plt.figure()
plt.plot(df['x'], label='X Axis')
plt.title('Gyroscope X Axis')
plt.xlabel('Time (samples)')
plt.ylabel('X Value')
plt.savefig('gyroscope_x_axis.png')  # Save the figure
plt.show()

# Plot 'y' variable
plt.figure()
plt.plot(df['y'], label='Y Axis', color='orange')
plt.title('Gyroscope Y Axis')
plt.xlabel('Time (samples)')
plt.ylabel('Y Value')
plt.savefig('gyroscope_y_axis.png')  # Save the figure
plt.show()

# Plot 'z' variable
plt.figure()
plt.plot(df['z'], label='Z Axis', color='green')
plt.title('Gyroscope Z Axis')
plt.xlabel('Time (samples)')
plt.ylabel('Z Value')
plt.savefig('gyroscope_z_axis.png')  # Save the figure
plt.show()
```

```python
plt.figure()
plt.plot(df['x'], label='X Axis')
plt.plot(df['y'], label='Y Axis', color='orange')
plt.plot(df['z'], label='Z Axis', color='green')
plt.title('Gyroscope X, Y, Z Axes')
plt.xlabel('Time (samples)')
plt.ylabel('Value')
plt.legend()
plt.savefig('gyroscope_xyz_combined.png')  # Save the combined figure
plt.show()
```
