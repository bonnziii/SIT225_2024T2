```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file, skipping the header if necessary
df = pd.read_csv('ultrasonic_data.csv', skiprows=1, header=None, names=['Timestamp', 'Distance'])

# Convert the Timestamp column to datetime format for better plotting
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y%m%d%H%M%S', errors='coerce')

# Drop any rows where the Timestamp conversion failed
df = df.dropna(subset=['Timestamp'])

# Plotting the Distance data
plt.figure(figsize=(10, 6))
plt.plot(df['Timestamp'], df['Distance'], marker='o')
plt.title('Ultrasonic Sensor Distance Over Time')
plt.xlabel('Time')
plt.ylabel('Distance (cm)')
plt.grid(True)
plt.show()
```
