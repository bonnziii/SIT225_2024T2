```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the CSV file into a DataFrame (replace 'your_data.csv' with your actual file name)
data = pd.read_csv('dht22_data.csv')

# Check the first few rows to confirm the data is loaded properly
print(data.head())

# Separate the independent (X) and dependent (y) variables
X = data['Temperature'].values.reshape(-1, 1)  # Temperature as X
y = data['Humidity'].values  # Humidity as y

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Print the model's parameters (slope and intercept)
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")

# Find the min and max temperature values
min_temp = data['Temperature'].min()
max_temp = data['Temperature'].max()

# Generate 100 equally spaced temperature values between the min and max of the dataset
test_temps = np.linspace(min_temp, max_temp, 100).reshape(-1, 1)

# Predict the humidity for these test temperatures
predicted_humidity = model.predict(test_temps)

# Print the test temperatures and corresponding predicted humidity values
for temp, humidity in zip(test_temps, predicted_humidity):
    print(f"Temperature: {temp[0]:.2f}, Predicted Humidity: {humidity:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the original data (replace with your correct file path if necessary)
data = pd.read_csv('dht22_data.csv')

# Separate the independent variable (Temperature) and dependent variable (Humidity)
X = data['Temperature'].values.reshape(-1, 1)
y = data['Humidity'].values

# Train the Linear Regression model using the original data
model = LinearRegression()
model.fit(X, y)

# Generate test temperature values (100 points between min and max of the original data)
min_temp = X.min()
max_temp = X.max()
test_temps = np.linspace(min_temp, max_temp, 100).reshape(-1, 1)

# Predict humidity values using the linear regression model
predicted_humidity = model.predict(test_temps)

# Calculate R-squared value for goodness-of-fit
r_squared = model.score(X, y)
print(f"R-squared for original data (dht22_data.csv): {r_squared}")

# Plot the original data points
plt.scatter(data['Temperature'], data['Humidity'], color='blue', label='Original Data (dht22_data.csv)')

# Plot the regression line
plt.plot(test_temps, predicted_humidity, color='red', label='Regression Line')

plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.legend()
plt.title('Linear Regression: Temperature vs Humidity (dht22_data.csv)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the manually filtered dataset (replace 'filtered_data.csv' with the correct filename)
filtered_data = pd.read_csv('filtered_data.csv')

# Separate the independent variable (Temperature) and dependent variable (Humidity)
X_filtered = filtered_data['Temperature'].values.reshape(-1, 1)
y_filtered = filtered_data['Humidity'].values

# Train the Linear Regression model using the manually filtered data
model_filtered = LinearRegression()
model_filtered.fit(X_filtered, y_filtered)

# Generate test temperature values (100 points between min and max of the filtered data)
min_temp = X_filtered.min()
max_temp = X_filtered.max()
test_temps = np.linspace(min_temp, max_temp, 100).reshape(-1, 1)

# Predict humidity values using the linear regression model
predicted_filtered_humidity = model_filtered.predict(test_temps)

# Plot the manually filtered data points and the regression line
plt.scatter(filtered_data['Temperature'], filtered_data['Humidity'], color='blue', label='Filtered Data')
plt.plot(test_temps, predicted_filtered_humidity, color='red', label='Regression Line')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Linear Regression: Temperature vs Humidity (Filtered Data)')
plt.legend()
plt.show()

# Check the R-squared value for goodness-of-fit
r_squared_filtered = model_filtered.score(X_filtered, y_filtered)
print(f"R-squared for manually filtered data: {r_squared_filtered}")
```
