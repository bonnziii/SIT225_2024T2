```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('DHT22_data_20240826225602.csv') 

# Convert the 'Timestamp' column to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'], format='%Y%m%d%H%M%S')

# Plot Temperature and Humidity on the same graph
plt.figure(figsize=(12, 6))  # Create a figure with a custom size

# Plot Temperature
plt.plot(data['Timestamp'], data['Temperature'], label='Temperature (°C)', color='red', marker='o')

# Plot Humidity
plt.plot(data['Timestamp'], data['Humidity'], label='Humidity (%)', color='blue', marker='x')

# Customize the plot
plt.xlabel('Time')  # X-axis label
plt.ylabel('Values')  # Y-axis label
plt.title('Temperature and Humidity Over Time')  # Title of the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()  # Display legend
plt.grid(True)  # Add grid lines to the plot

# Display the plot
plt.show()
```
