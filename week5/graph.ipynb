```python
import matplotlib.pyplot as plt
import pandas as pd  # Add this line to import Pandas

# Read cleaned CSV
df = pd.read_csv('gyroscope_data.csv')

# Convert timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot x, y, z separately
plt.subplot(3, 1, 1)
plt.plot(df['Timestamp'], df['X-axis'], label="X-axis")
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df['Timestamp'], df['Y-axis'], label="Y-axis")
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df['Timestamp'], df['Z-axis'], label="Z-axis")
plt.legend()

plt.tight_layout()
plt.show()

# Plot all variables in one plot
plt.figure(figsize=(10, 6))
plt.plot(df['Timestamp'], df['X-axis'], label="X-axis")
plt.plot(df['Timestamp'], df['Y-axis'], label="Y-axis")
plt.plot(df['Timestamp'], df['Z-axis'], label="Z-axis")
plt.legend()
plt.show()
```
